```python
#import plotly library 
import plotly
import plotly.graph_objects as go
import plotly.offline as py
import plotly.express as px
import pandas as pd 
```

```python
#read in dataset 
df = pd.read_csv('covid_data/country_vaccinations.csv')
df.head()
```

```python
#drop null values from total_vaccinations column 
df1=df[df['total_vaccinations'].notna()]
df1.head()
```

```python
#getting the most current number of total_vaccinations of each country
df2 = df1.groupby(['country']).max()
df2.head()
```

```python
#creating smaller data frame with just country & total vaccinations 
total_country_vax = df1[["country", "total_vaccinations"]]

#getting the most current number of total_vaccinations of each country
total_country_vax = total_country_vax.groupby(['country']).max()
total_country_vax.head()
```

```python
#check columns in our data frame 
total_country_vax.columns
```

```python
#specifying x & y axis and layout of bar graph 
data = [go.Bar(x=total_country_vax.index, y=total_country_vax.total_vaccinations, name='Total Vaccinations by Country')]
layout = go.Layout(title = 'Total Vaccinations by Country')
```

```python
#create our figure to be plotted
fig = go.Figure(data = data, layout = layout)
```

```python
#plot our figure
py.plot(fig)
```
